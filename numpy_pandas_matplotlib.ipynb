```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3]
y = [2, 4, 6]

plt.figure(figsize= (5,2),dpi=300)
plt.plot(x,y, label = '2x', color = 'red', linewidth = 4, marker = '*', markersize= 10, markeredgecolor = 'blue')
plt.title('The graph', fontdict={'fontname':'Comic Sans MS', 'fontsize':20 })
plt.ylabel('y axis',fontdict={'fontname':'Comic Sans MS', 'fontsize':20 })
plt.xlabel('x axis',fontdict={'fontname':'Comic Sans MS', 'fontsize':20 })
plt.xticks([1,2, 4,3])
plt.yticks(y)
z = [4, 6, 10] 
u = [7, 11, 5]
plt.plot(z,u, 'g.--', label = '3x', linewidth = 2, marker = '.', markersize= 10, markeredgecolor = 'blue')

x2 = np.arange(0,10,0.5)
plt.plot(x2[:5],2 *np.sin(x2[:5]), '--', label = 'sin(x2)', linewidth = 2)
plt.plot(x2[5:20],2 *np.sin(x2[5:20]),linewidth = 2)
plt.legend()
plt.savefig('thegraph.png', dpi = 300)
plt.show()
```

```python
labels = ['A', 'B', 'C']
values = [1, 4, 2]
plt.figure(figsize=(5,3), dpi = 300)
bars=plt.bar(labels, values)
'''bars[0].set_hatch('.')
bars[1].set_hatch('/')
bars[2].set_hatch('*')'''

patterns = ['.', '/', '*']
for bar in bars:
  bar.set_hatch(patterns.pop())
plt.show()
```

```python
gas = pd.read_csv('gas_prices.csv')
plt.figure(figsize=(10,4), dpi = 200)
plt.title('gas prices ', fontdict = {'fontname':'Arial', 'fontsize': 15})
'''plt.plot(gas.Year, gas.Canada,marker = '*', markeredgecolor = 'r', label = 'Canada')
plt.plot(gas['Year'], gas.France,label = 'France')
plt.plot(gas.Year, gas.Germany, label= 'Germany')'''
countries = ['USA', 'UK', 'France', 'Italy']
for gass in gas:
  if gass in countries:
    plt.plot(gas.Year,gas[gass], marker= '.',label = gass)
plt.xticks(gas['Year'])
plt.legend()
plt.savefig('gasprices.png', dpi = 200)
plt.show()
```
